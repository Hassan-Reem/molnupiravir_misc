```python
tree = toytree.tree("/Users/reem/sampled_tree.txt-subtree-0.nw")
mov_samples = "/Users/reem/matches.txt"
min_height = tree.ntips * 30  
height = 1000


# Color leaves: blue for default, red if in matches
node_colors = ["blue"] * tree.nnodes
with open(mov_samples, "r") as f:
    mov_samples = set(line.strip() for line in f)
for leaf in tree[:tree.ntips]:  
    if leaf.name in mov_samples:
        node_colors[leaf.idx] = "red"
    else:
        node_colors[leaf.idx] = "blue"

for node in tree:
    if node in tree[:tree.ntips]:
        node_markers = "o"
   



canvas, axes, marks = tree.draw(
    node_markers=node_markers,
    node_sizes=15,  
    node_mask=False, 
    node_colors=node_colors,
    scale_bar=True,
    width=1200,  
    height=height,
    edge_widths=2,  
    tip_labels_align=True)


canvas.style.update({"background-color": "white"})
```

```python
import toyplot.svg
toyplot.svg.render(canvas, "/Users/reem/Desktop/tree-plot.svg")
```
