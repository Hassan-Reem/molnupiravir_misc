```python
import pandas as pd
import bte
import numpy
import matplotlib as plt
import plotly.express as px
```

```python
sand_2022 = pd.read_csv("/Users/reem/sanderson_2022.tsv", sep="\t")
sand_2022.head(5)
```

```python
from collections import Counter

# Extract all TNCs from spectrum column
all_tncs = []
all_ga_tncs = []
all_ct_tncs = []

for spectrum_str in sand_2022['spectrum']:
    if pd.isna(spectrum_str):
        continue
    
    # Parse each mutation: A[G>A]A, T[G>A]G, etc.
    mutations = spectrum_str.split(',')
    for mut in mutations:
        all_tncs.append(mut.strip())
        
        if 'G>A' in mut:
            all_ga_tncs.append(mut.strip())
        elif 'C>T' in mut:
            all_ct_tncs.append(mut.strip())

# Count frequencies
tnc_counter = Counter(all_tncs)
ga_counter = Counter(all_ga_tncs)
ct_counter = Counter(all_ct_tncs)

print("=== TOP 10 G>A CONTEXTS ===")
for tnc, count in ga_counter.most_common(10):
    print(f"{tnc}: {count}")

print("\n=== TOP 10 C>T CONTEXTS ===")
for tnc, count in ct_counter.most_common(10):
    print(f"{tnc}: {count}")

print(f"\nTotal G>A mutations: {sum(ga_counter.values())}")
print(f"Total C>T mutations: {sum(ct_counter.values())}")
```

```python
fig = px.bar(x=list(ga_counter.keys()), y=list(ga_counter.values()),
             title='G>A Mutation Contexts in Sanderson 2022 MOV Nodes')
fig.update_layout(xaxis_title='Trinucleotide Context', yaxis_title='Count')
fig.show()
```

```python
tree = bte.MATree("/Users/reem/2026_updated_tree.pb")
```

```python
from scipy.spatial.distance import pdist, squareform
import numpy as np

# Get MOV nodes


def identify_clusters(nodes_df, tree, max_distance=0.02):
    """
    Find clusters of closely related MOV nodes.
    max_distance = genetic distance threshold (tweak as needed)
    """
    clusters = []
    visited = set()
    
    for idx, node_id in enumerate(nodes_df['node_id']):
        if node_id in visited:
            continue
        
        # Find all nodes within max_distance of this node
        cluster = [node_id]
        visited.add(node_id)
        
        # Get descendants and nearby nodes from tree
        node = tree.get_node(node_id)
        if node is None:
            continue
        
        # Simple approach: find all nodes in subtree
        # (you may want to use branch_length or genetic distance instead)
        for other_idx, other_node_id in enumerate(nodes_df['node_id']):
            if other_node_id in visited:
                continue
            
            # Check if nodes are related (within same clade)
            try:
                mrca = tree.get_mrca([node_id, other_node_id])
                if mrca and mrca.branch_length < max_distance:
                    cluster.append(other_node_id)
                    visited.add(other_node_id)
            except:
                pass
        
        if len(cluster) > 1:  # Only keep multi-node clusters
            clusters.append(cluster)
    
    return clusters

# Run it
mov_clusters = identify_clusters(sand_2022, tree)
multi_person_clusters = [c for c in mov_clusters if len(c) > 1]

print(f"Total MOV clusters: {len(mov_clusters)}")
print(f"Multi-node clusters: {len(multi_person_clusters)}")
print(f"Nodes in clusters: {sum(len(c) for c in multi_person_clusters)}")
print(f"Largest cluster: {max(len(c) for c in mov_clusters) if mov_clusters else 0}")

# Show top clusters
sorted_clusters = sorted(multi_person_clusters, key=len, reverse=True)
print("\n=== TOP 10 CLUSTERS ===")
for i, cluster in enumerate(sorted_clusters[:10]):
    print(f"Cluster {i+1}: {len(cluster)} nodes")
    # Show details
    cluster_nodes = sand_2022[sand_2022['node_id'].isin(cluster)]
    print(f"  Countries: {cluster_nodes['country'].value_counts().to_dict()}")
    print(f"  Dates: {cluster_nodes['date'].min():.0f} to {cluster_nodes['date'].max():.0f}")
```
