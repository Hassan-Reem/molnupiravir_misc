```python
import plotly.express as px 
import pandas as pd
```

```python
# Read in mov means
df = pd.read_csv("/Users/reem/GtoAprobs_new.tsv", delimiter="\t")
df.head()
```

```python
import plotly.graph_objects as go
# Plot proprtions of G>A mutations fro MOV vs Normal samples

fig = fig = go.Figure(data=[go.Bar(name='Molnupiravir', x=df['Mutational_Context'], y=df['Molnupiravir']), go.Bar(name='Normal', x=df['Mutational_Context'], y=df['Normal'])], 
                      layout=go.Layout(title='G-to-A mutation spectrum by trinucleotide context', yaxis=dict(range=[0,0.3])))

fig.update_layout(margin=dict(l=20, r=20, t=50, b=20), legend=dict(x=0.8, y=0.95))
fig.write_image("/Users/reem/molnupiravir_paper/Figures/mov_vs_normal_ga_proportions.pdf")
fig.show()
```

```python
ctprobs = pd.read_csv("/Users/reem/CtoTprobs_new.tsv", delimiter="\t")
ctprobs.head()
```

```python
# Plot proprtions of C-to-T mutations fro MOV vs Normal samples

fig = fig = go.Figure(data=[go.Bar(name='Molnupiravir', x=ctprobs['Mutational_Context'], y=ctprobs['Molnupiravir']), go.Bar(name='Normal', x=ctprobs['Mutational_Context'], y=ctprobs['Normal'])], 
                      layout=go.Layout(title='C-to-T mutation spectrum by trinucleotide context', yaxis=dict(range=[0,0.3])))

fig.update_layout(margin=dict(l=20, r=20, t=50, b=20), legend=dict(x=0.8, y=0.95))
fig.write_image("/Users/reem/molnupiravir_paper/Figures/mov_vs_normal_ct_proportions.pdf")
fig.show()
```

```python
fig = px.bar(df, x='Mutational_Context', y='Normal',
    title='Non-MOV mean G>A proportions by mutation context')
fig.layout.yaxis.range=[0,0.3]
fig.show()
```

```python
# Plot the last row in the Mov_dataframe "means_row" vs GtoA context
context_cols = [col for col in df.columns if col not in ["seqName", "LLR","Unnamed: 0"]]
mean_row = df[df['seqName'] == 'Mean'].iloc[0]
#print(mean_row)
plot_df = mean_row[context_cols].reset_index()
plot_df.columns = ['Context', 'Mean']
fig = px.bar(plot_df, x='Context', y='Mean',
    title='Mean Values of Mutation Contexts')
fig.show()
```
