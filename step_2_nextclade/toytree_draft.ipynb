```python
import toytree
import toyplot
```

```python
tree = toytree.tree("/Users/reem/nwk_files/sampled_tree.txt-subtree-3.nw")
mov_samples = "/Users/reem/matches.txt"
min_height = tree.ntips * 30  
height = max(min_height, 1200)


# Color leaves: blue for default, red if in matches
node_colors = ["blue"] * tree.nnodes
with open(mov_samples, "r") as f:
    mov_samples = set(line.strip() for line in f)
for leaf in tree[:tree.ntips]:  
    if leaf.name in mov_samples:
        node_colors[leaf.idx] = "red"
    else:
        node_colors[leaf.idx] = "blue"

canvas, axes, marks = tree.draw(
    node_sizes=15,  
    node_mask=False, 
    node_colors=node_colors,
    scale_bar=True,
    width=1000,  
    height=height,
    edge_widths=2,  
    tip_labels_align=True)


canvas.style.update({"background-color": "white"})
```

```python
#Try again to get a less crowded tree

newick_path= "/Users/reem/nwk_files/sampled_tree.txt-subtree-3.nw"

mov_samples = "/Users/reem/matches.txt"
with open(mov_samples) as f:
    mov_samples = [line.strip() for line in f]
tree = toytree.tree(newick_path)


min_height = tree.ntips * 45 
height = max(min_height, 1500)


node_colors = ["none"] * tree.nnodes

# Color ONLY leaves: blue for default, red if in matches
for leaf in tree[:tree.ntips]:
    if leaf.name in mov_samples:
        node_colors[leaf.idx] = "red"
    else:
        node_colors[leaf.idx] = "blue"

canvas, axes, marks = tree.draw(
    node_sizes=12,  
    node_mask=False, 
    node_colors=node_colors,
    scale_bar=True,
    width=1200,  
    height=height,
    edge_widths=1.5,
    tip_labels_align=True,
    tip_labels_style={
        "font-size": "10px",
        "-toyplot-anchor-shift": "15px"  # Adds space between node and label
    },
    node_style={
        "stroke": "white",
        "stroke-width": 1.5
    },
    edge_style={
        "stroke": "#555555",
        "stroke-opacity": 0.7
    },
    edge_align_style={
        "stroke": "lightgray", 
        "stroke-width": 0.5,
        "stroke-dasharray": "2,2"
    },
    padding=60,  
    layout='r'  
)

canvas.style.update({"background-color": "white"})
```
