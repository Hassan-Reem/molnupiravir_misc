```python
import pandas as pd
import plotly.express as px
```

Notebook visualizing estimated mutation class probability distributions for MOV-associated versus normal samples.

```python
df = pd.read_csv("/Users/reem/Downloads/estimated_mutation_distribution.tsv", sep="\t")
df.head()
```

```python
fig = px.bar(df, x='MutationType', y="Molnupiravir", title='Estimated Mutation Distribution under Molnupiravir Treatment',
    category_orders={
        "MutationType": ["A→C", "A→G", "A→T", "C→A", "C→G", "C→G", "C→T", "G→A", "G→C", "G→T", "T→A", "T→C"]
    }
)

fig.show()
```

```python
fig = px.bar(df, x='MutationType', y="Normal", title='Estimated Mutation Distribution under Normal Conditions',
    category_orders={
        "MutationType": ["A→C", "A→G", "A→T", "C→A", "C→G", "C→G", "C→T", "G→A", "G→C", "G→T", "T→A", "T→C"]
    },
)
fig.update_yaxes(range =[0,0.4])

fig.show()
```
